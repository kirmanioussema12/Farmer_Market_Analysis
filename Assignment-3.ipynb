# <span style="color:Green">P1: Load the dataset. How many observations are recorded in the dataset? What are the data types of each column? Are there missing values in any columns? </span>

```python
#We transform the excel file to a csv file using the pandas library  in order to be observable 
#And then we load it in a dataframe to be able to manipulate it and perform the necessary analysis on it.
import pandas as pd 
import numpy as np
#df = pd.read_excel(r"C:\\Users\\MSI\\Desktop\\Farmer Project\\Farmers' Market Observation-dataset.xlsx").to_csv(r"C:\\Users\\MSI\\Desktop\\Farmer Project\\Farmers' Market Observation-dataset.csv", index=False)
df = pd.DataFrame(pd.read_csv(r"C:\\Users\\MSI\\Desktop\\Farmer Project\\Farmers' Market Observation-dataset.csv"))
#We check number of records, missing values and data types of each column 
#using  these functions df.shape, df.isna().sum() and df.dtypes 
print("Number of Records in the dataset of the Farmers:", df.shape[0])
print("Number of Missing Values in each Column:", df.isna().sum())
print("Data Types of each Column:", df.dtypes)
```

# <span style="color:Green">P2: How many unique cities are represented in the dataset? Which cities show the highest average handwashing attempts?</span>

```python
#Using the unique function to find the number and the list of unique cities in the dataset
print(" The  Number and the List of Unique Cities in the dataset:", len(df["City"].unique()), "\n", df["City"].unique())
#Using the groubpy method to groub cities per handwashing station and then we calculate the mean of the handwashing station in each city
#and sort them in descending order to find the cities with the highest average handwashing attempts
print("The Cities with the highest average handwashing attempts are:","\n",df.groupby("City")["Visible_handwashing_station"].mean().sort_values(ascending=False))
```

# <span style="color:Green">P3: How many vendors were observed in total (vendor_ID)? What is the average number of vendors per city (vendor_ID)? What is the distribution of vendor roles?</span>

```python
#We check if there are any duplicated and missing vendor_ID using the duplicated() and isna() functions respectively
print("Duplicated vendor_ID:", df["vendor_ID"].duplicated().sum())
print("Missing vendor_ID:", df["vendor_ID"].isna().sum())
total_vendors = df["vendor_ID"].nunique()
#We calculate the total number of vendors observed and the average number of vendors per city using the nunique() function
print("Total vendors observed:", total_vendors)
avg_vendors = df.groupby("City")["vendor_ID"].count().mean()
print("Average number of vendors per city:", avg_vendors)
# Distribution of vendor roles using the value_counts() function
print(df["vendor_roles"].value_counts())
```

# <span style="color:Green">P4: How many markets operate year-round? What is the distribution of market visits by month?</span>

```python
#We group the data by market_ID and check the first value of the Yearround column 
# to determine if the market operates year-round or not.
markets=df.groupby("market_ID")["Yearround"].first()
# Count markets with 'Yes'
yearround_markets = (markets == "Yes").sum()
print("Number of markets that operate year-round:", yearround_markets)
#We group the data by market_ID and check the first value of the Visit_month column   
#to find out the distribution of markets by visit month.
markets_months=df.groupby("market_ID")["Visit_month"].first().value_counts()
print("Distribution of markets by visit month:\n", markets_months)

```

# <span style="color:Green">PBonus: Write Python scripts to create Tables 1-4 as they appear in the research paper. </span>

```python
#Codes of PBonus here
import pandas as pd

# -------------------------------------------------
# STEP 1: Create a market-level dataset (1 row per market)
# -------------------------------------------------
markets = df.groupby("market_ID").first()

total_markets = len(markets)

# -------------------------------------------------
# Function to compute counts and percentages
# -------------------------------------------------
def create_table(column):
    counts = markets[column].value_counts()
    percentages = (counts / total_markets) * 100
    table = pd.DataFrame({
        "No.": counts,
        "%": percentages.round(1)
    })
    return table


# -------------------------------------------------
# TABLE 1 SECTIONS
# -------------------------------------------------

# Area
table_area = create_table("City")

# Operation schedule
table_schedule = create_table("Yearround")

# Location
table_location = create_table("Location")

# Setting
#table_setting = create_table("Setting")

# Month of visit
table_month = create_table("Visit_month")

# Day of visit
table_day = create_table("Visit_day")

# Communal/shared handwashing sinks
table_sink = create_table("communal_sinks")

# Electrical access
table_electric = create_table("electrical_access")

# Signs prohibiting pets
table_signs = create_table("pet_signs")

# Pets visible at market
table_pets = create_table("pets_visible")


# -------------------------------------------------
# PRINT RESULTS
# -------------------------------------------------

print("Table 1: Characteristics of Farmers’ Markets\n")

print("Area:")
print(table_area, "\n")

print("Operation schedule:")
print(table_schedule, "\n")

print("Location:")
print(table_location, "\n")

print("Setting:")
#print(table_setting, "\n")

print("Month of visit:")
print(table_month, "\n")

print("Day of week of visit:")
print(table_day, "\n")

print("Communal or shared handwashing sinks:")
print(table_sink, "\n")

print("Electrical access:")
print(table_electric, "\n")

print("Signs prohibiting animal/pet access onsite:")
print(table_signs, "\n")

print("Pets visible onsite at the market:")
print(table_pets)
```
